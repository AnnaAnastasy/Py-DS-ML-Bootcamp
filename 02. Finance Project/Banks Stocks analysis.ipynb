**Get the stock data from Jan 1st 2006 to Jan 1st 2022 for each of these banks. Set each bank to be a separate dataframe, with the variable name for that bank being its ticker symbol. This will involve a few steps:**
1. Use datetime to set start and end datetime objects.
2. Choose correct ticker symbol for each bank.
2. Use datareader to grab info on the stock.

**Use [this documentation page](https://pandas-datareader.readthedocs.io/en/latest/remote_data.html) for hints and instructions:**
    
    # Bank of America
    BAC = data.DataReader("BAC", 'google', start, end)

### WARNING: MAKE SURE TO CHECK THE LINK ABOVE FOR THE LATEST WORKING API. "google" does not always work. So in this project I am using "yahoo" instead.

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2022,1,1)
```

```python
BAC = data.DataReader("BAC", 'yahoo', start, end)
C = data.DataReader("C", 'yahoo', start, end)
GS = data.DataReader("GS", 'yahoo', start, end)
JPM = data.DataReader("JPM", 'yahoo', start, end)
MS = data.DataReader("MS", 'yahoo', start, end)
WF = data.DataReader("WF", 'yahoo', start, end)
```

**Create a list of the ticker symbols (as strings) in alphabetical order.**

```python
tickers = ['BAC','C','GS','JPM','MS','WF']
```

**Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WF], axis=1, keys=tickers)
bank_stocks
```

**Set the column name levels:**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

**Check the head and tail of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

```python
bank_stocks.tail()
```

# EDA

Let's explore the data a bit! 
**What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs(key='Close',axis=1, level='Stock Info').max()
```

**Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

**We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick +' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

**Create a pairplot using seaborn of the returns dataframe.**

```python
import seaborn as sns
sns.set_style('whitegrid')
```

```python
sns.pairplot(data=returns)
```

Here you notice that the standard deviation is actually quite a bit more stretched out especially at point 0.6. And if we look back for a normal year 2021 for a bank WF point 0.06 is 10 times smaller than the maximum value for a Citigroup at 2008.

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

**Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
plt.figure(figsize=(14,4))
BAC.loc[pd.IndexSlice['2008-01-01': '2008-12-31'],'Close'].rolling(30).mean().plot(label='30 Days Average')
BAC.loc[pd.IndexSlice['2008-01-01': '2008-12-31'],'Close'].plot(label='BAC Close')
plt.legend()
```

**Create a heatmap of the correlation between the stocks Close Price.**

```python
close_corr = bank_stocks.xs(key='Close',axis=1, level='Stock Info').corr()
sns.heatmap(close_corr, linewidths=1, cmap="YlGnBu", annot=True)
```

```python
close_corr.iplot(kind='heatmap', colorscale="rdylbu")
```

**Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(close_corr, annot=True)
```

**Use .iplot(kind='candle) to create a candle plot of Bank of America's stock from Jan 1st 2008 to Jan 1st 2010.**

```python
BAC.loc[pd.IndexSlice['2008-01-01':'2010-01-01'],['Open','High','Low','Close']].iplot(kind='candle')
```

**Use .ta_plot(study='sma') to create a Simple Moving Averages plot of Morgan Stanley for the years 2008 and 2009.**

```python
MS.loc[pd.IndexSlice['2008-01-01': '2009-12-31'],'Close'].ta_plot(study='sma', periods=[7,30])
```

**Use .ta_plot(study='boll') to create a Bollinger Band Plot for Bank of America for the years 2008 and 2009.**

```python
BAC.loc[pd.IndexSlice['2008-01-01': '2009-12-31'],'Close'].ta_plot(study='boll')
```
